*Importando as bibliotecas necessárias*

```python
import numpy as np
import sympy as sp
```

*Chamamos o lado do quadrado de "L", a diagonal do quadrado "D", os catetos do triângulo de "l" e a hipotenusa do triângulo de "h".*

```python
L, D, l, h = sp.symbols("L D l h")
```

*O lado do quadrado "L" é o mesmo do raio do setor circular. Enquanto a diagonal é igual ao raio do setor circular (ou lado do quadrado) mais a hipotenusa do triângulo. Também sabemos que a diagonal do quadrado é igual ao lado vezes raiz de 2.*

$D = L + h = L \times \sqrt{2}$

```python
eq1 = sp.Eq(D, L + h)
eq2 = sp.Eq(D, L * np.sqrt(2))
sols1 = sp.solve([eq1, eq2], [D, h])
```

```python
sols1[D]
```

```python
sols1[h]
```

*Como esse triângulo demarcado possui sua hipotenusa como diagonal de um quadrado, seus ângulos são 45°, 45° e 90°. Logo, seus lados são a hipotenusa dividido por raiz de 2.*

$h = l \times \sqrt{2}$

```python
eq3 = sp.Eq(h, l * np.sqrt(2))
sols2 = sp.solve(eq3, l)
```

```python
sols2[0]
```

```python
cat = sols2[0].subs({h : sols1[h]})
```

*A área de um triângulo retângulo pode ser definida como o produto dos catetos dividido por 2.*

$A_{\Delta} = \frac{l_{a} \times l_{b}}{2}$

```python
area = (cat * cat)/2
```
