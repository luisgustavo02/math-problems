*Importando as bibliotecas necessrias*

```python
import numpy as np
import sympy as sp
```

*Chamamos o lado do quadrado de "L", a diagonal do quadrado "D", os catetos do tringulo de "l" e a hipotenusa do tringulo de "h".*

```python
L, D, l, h = sp.symbols("L D l h")
```

*O lado do quadrado "L"  o mesmo do raio do setor circular. Enquanto a diagonal  igual ao raio do setor circular (ou lado do quadrado) mais a hipotenusa do tringulo. Tambm sabemos que a diagonal do quadrado  igual ao lado vezes raiz de 2.*

$D = L + h = L \times \sqrt{2}$

```python
eq1 = sp.Eq(D, L + h)
eq2 = sp.Eq(D, L * np.sqrt(2))
sols1 = sp.solve([eq1, eq2], [D, h])
```

```python
sols1[D]
```

```python
sols1[h]
```

*Como esse tringulo demarcado possui sua hipotenusa como diagonal de um quadrado, seus ngulos so 45, 45 e 90. Logo, seus lados so a hipotenusa dividido por raiz de 2.*

$h = l \times \sqrt{2}$

```python
eq3 = sp.Eq(h, l * np.sqrt(2))
sols2 = sp.solve(eq3, l)
```

```python
sols2[0]
```

```python
cat = sols2[0].subs({h : sols1[h]})
```

*A rea de um tringulo retngulo pode ser definida como o produto dos catetos dividido por 2.*

$A_{\Delta} = \frac{l_{a} \times l_{b}}{2}$

```python
area = (cat * cat)/2
```
